```python
import requests

# Fetch a web page
r = requests.get("https://medium.com/@aghadurrezkhan75/intellifarm-farm-supervision-productivity-analysis-platform-b81961e79210")
print(r.text)
```

```python
import re

# Remove HTML tags using RegularExpression
pattern = re.compile(r'<.*?>')  # tags look like <...>
print(pattern.sub('', r.text))  # replace them with blank
```

```python
from bs4 import BeautifulSoup
```

```python



# Remove HTML tags using Beautiful Soup library
soup = BeautifulSoup(r.text, "html5lib")
print(soup.get_text())
```

```python

#mw-content-container
# Find all articles
summaries = soup.find_all("div", class_="l")
summaries[0]
```

```python
scraped_data = summaries[0].get_text().strip()
```

```python
scraped_data= scraped_data.lower()
```

```python
import re

# Remove punctuation characters
scraped_data = re.sub(r"[^a-zA-Z0-9]", " ", scraped_data)
print(scraped_data)
```

```python

# Split text into tokens (words)
words = scraped_data.split()
print(words)
```

```python
import os
import nltk
nltk.data.path.append(os.path.join(os.getcwd(), "nltk_data"))
```

```python

from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')

# Split text into words using NLTK
words = word_tokenize(scraped_data)
print(words)
```
